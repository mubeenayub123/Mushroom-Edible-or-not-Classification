# Machine Learning Cycle

### Four phases of a Machine Learning Cycle are

### Training Phase

    Build the Model using Training Data

### Testing Phase

     Evaluate the performance of Model using Testing Data

### Application Phase

     Deploy the Model in the Real-world, to predict Real-time unseen Data

### Feedback Phase

    Take Feedback from the Users and Domain Experts to improve the Model

# Step 5: Label Encoding the Sample Data (Input and Output is converted in Numeric Representation)

## Step 5.1: Train the Label Encoder

```python
# Train the Label Encoder

''' 
*------------------ TRAIN_LABEL_ENCODER --------------------*
|        Function: Fit()                                    |
|              Purpose: Fit or Train the Label Encoder      |
|        Arguments:                                         |
|               Labels: Target Values                       |
|        Return:                                            |
|               Instance: Returns an instance of self       |
*-----------------------------------------------------------*
''' 

# Labels

capshape = pd.DataFrame({"cap-shape":["b","c","x","f","k","s"]})
capsurface = pd.DataFrame({"cap-surface":["f","g","y","s"]})
capcolor = pd.DataFrame({"cap-color":["n","b","c","g","r","p","u","e","w","y"]})
odor = pd.DataFrame({"odor":["a","l","c","y","f","m","n","p","s"]})
gillsize = pd.DataFrame({"gill-size":["b","n"]})
gillcolor = pd.DataFrame({"gill-color":["k","n","b","h","g","r","o","p","u","e","w","y"]})
stalkroot = pd.DataFrame({"stalk-root":["b","c","u","e","z","r","o"]})
sporeprintcolor=pd.DataFrame({"spore-print-color":["k","n","b","h","r","o","u","w","y"]})
population=pd.DataFrame({"population":["a","c","n","s","v","y"]})
habitat=pd.DataFrame({"habitat":["g","l","m","p","u","w","d"]})
Output=pd.DataFrame({"Output":["e","p"]})  

# Initialize the Label Encoders 
capshape_label_encoder = LabelEncoder()
capsurface_label_encoder = LabelEncoder()
capcolor_label_encoder = LabelEncoder()
odor_label_encoder = LabelEncoder()
gillsize_label_encoder = LabelEncoder()
gillcolor_label_encoder = LabelEncoder()
stalkroot_label_encoder = LabelEncoder()
sporeprintcolor_label_encoder = LabelEncoder()
population_label_encoder = LabelEncoder()
habitat_label_encoder = LabelEncoder()
Output_label_encoder = LabelEncoder()

# Train the Label Encoders

capshape_label_encoder.fit(np.ravel(capshape))
capsurface_label_encoder.fit(np.ravel(capsurface))
capcolor_label_encoder.fit(np.ravel(capcolor))
odor_label_encoder.fit(np.ravel(odor))
gillsize_label_encoder.fit(np.ravel(gillsize))
gillcolor_label_encoder.fit(np.ravel(gillcolor))
stalkroot_label_encoder.fit(np.ravel(stalkroot))
sporeprintcolor_label_encoder.fit(np.ravel(sporeprintcolor))
population_label_encoder.fit(np.ravel(population))
habitat_label_encoder.fit(np.ravel(habitat))
Output_label_encoder.fit(np.ravel(Output))
```

## Step 5.2: Label Encode the Output

```python
# Label Encoding of the Output

''' 
*------------------ LABEL_ENCODE_OUTPUT --------------------*
|        Function: Transform()                              |
|              Purpose: Transform Input (Categorical)       |
|                       into Numerical Representation       |
|        Arguments:                                         |
|              Attribute: Target values                     |
|        Return:                                            |
|              Attribute: Numerical Representation          |
*-----------------------------------------------------------*
'''

sample_data_encoded_output = sample_data.copy()
original_sample_data = sample_data.copy()

# Transform Output of into Numerical Representation

print("\n\nOutput Attribute After Label Encoding:")
print("========================================\n")
sample_data["encoded_Output"] = Output_label_encoder.transform(sample_data['Output'])
print(sample_data[["Output", "encoded_Output"]])

# Print Original and Encoded Ouput Sample Data

sample_data_encoded_output[['cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-size','gill-color', 'stalk-root', 
'spore-print-color', 'population','habitat', 'Output']] = sample_data[['cap-shape', 'cap-surface', 'cap-color', 
'odor', 'gill-size','gill-color', 'stalk-root', 'spore-print-color', 'population','habitat', 'encoded_Output']]
pd.set_option("display.max_rows", None, "display.max_columns", None)
print("\n\nOriginal Sample Data:")
print("=====================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(original_sample_data)
print("\n\nSample Data after Label Encoding of Output:")
print("===========================================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output)

# Save the Transformed Features into CSV File 

sample_data_encoded_output.to_csv(r'sample-data-encoded-output.csv', index = False, header = True)
```

## Step 5.3: Label Encode the Input

```python
# Label Encoding of the Input

''' 
*------------------- LABEL_ENCODE_INPUT --------------------*
|        Function: Transform()                              |
|              Purpose: Transform Input (Categorical)       |
|                          into Numerical Representation    |
|        Arguments:                                         |
|              Attribute: Target values                     |
|        Return:                                            |
|              Attribute: Numerical Representation          |
*-----------------------------------------------------------*
'''

sample_data_encoded = sample_data_encoded_output.copy()
sample_data_encoded_output_orignal = sample_data_encoded_output.copy()

# Transform Input Attributes into Numerical Representation

print("\n\nCap-shape Attribute After Label Encoding:")
print("======================================\n")
sample_data_encoded_output["encoded_cap-shape"] = capshape_label_encoder.transform(sample_data_encoded_output['cap-shape'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["cap-shape", "encoded_cap-shape"]])

print("\n\nCap-surface Attribute After Label Encoding:")
print("======================================\n")
sample_data_encoded_output["encoded_cap-surface"] = capsurface_label_encoder.transform(sample_data_encoded_output['cap-surface'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["cap-surface", "encoded_cap-surface"]])

print("\n\nCap-color Attribute After Label Encoding:")
print("=======================================\n")
sample_data_encoded_output["encoded_cap-color"] = capcolor_label_encoder.transform(sample_data_encoded_output['cap-color'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["cap-color", "encoded_cap-color"]])

print("\n\nOdor Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_odor"] = odor_label_encoder.transform(sample_data_encoded_output['odor'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["odor", "encoded_odor"]])

print("\n\nGill-size Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_gill-size"] = gillsize_label_encoder.transform(sample_data_encoded_output['gill-size'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["gill-size", "encoded_gill-size"]])

print("\n\nGill-color Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_gill-color"] = gillcolor_label_encoder.transform(sample_data_encoded_output['gill-color'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["gill-color", "encoded_gill-color"]])

print("\n\nStalk-root Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_stalk-root"] = stalkroot_label_encoder.transform(sample_data_encoded_output['stalk-root'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["stalk-root", "encoded_stalk-root"]])

print("\n\nSpore-print-color Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_spore-print-color"] = sporeprintcolor_label_encoder.transform(sample_data_encoded_output['spore-print-color'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["spore-print-color", "encoded_spore-print-color"]])

print("\n\nPopulation Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_population"] = population_label_encoder.transform(sample_data_encoded_output['population'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["population", "encoded_population"]])

print("\n\nHabitat Attribute After Label Encoding:")
print("========================================\n")
sample_data_encoded_output["encoded_habitat"] = habitat_label_encoder.transform(sample_data_encoded_output['habitat'])
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded_output[["habitat", "encoded_habitat"]])

# Print Original and Encoded Sample Data

sample_data_encoded[['cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-size','gill-color', 'stalk-root', 
'spore-print-color', 'population','habitat', 'Output']] = sample_data_encoded_output[['encoded_cap-shape', 'encoded_cap-surface', 
'encoded_cap-color', 'encoded_odor', 'encoded_gill-size','encoded_gill-color', 'encoded_stalk-root', 
'encoded_spore-print-color', 'encoded_population','encoded_habitat', 'Output']]
print("\n\nOriginal Sample Data:")
print("=====================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(original_sample_data)
print("\n\nSample Data after Label Encoding:")
print("=================================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded)

# Save the Transformed Features into CSV File 

sample_data_encoded.to_csv(r'sample-data-encoded.csv', index = False, header = True)
```

# Step 6: Execute the Training Phase 

## Step 6.1: Splitting Sample Data into Training Data and Testing Data

```python
# Splitting Sample Data into Training Data and Testing Data

''' 
*------------------- SPLIT_SAMPLE_DATA ---------------------*
|        Function: train_test_split()                       |
|              Purpose: Split arrays or matrices into       |
|                       random train and test subsets       |
|        Arguments:                                         |
|              arrays: sequence of indexables               |
|              test_size: float or int                      |
|        Return:                                            |
|              splitting: list                              |
*-----------------------------------------------------------*
'''

training_data_encoded, testing_data_encoded = train_test_split( sample_data_encoded , test_size=0.2 , random_state=0 , shuffle = False)

# Save the Training and Testing Data into CSV File 

training_data_encoded.to_csv(r'training-data-encoded.csv', index = False, header = True)
testing_data_encoded.to_csv(r'testing-data-encoded.csv', index = False, header = True)

# print Training and Testing Data

print("\n\nTraining Data:")
print("==============\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(training_data_encoded)
print("\n\nTesting Data:")
print("==============\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(testing_data_encoded)
```

## Step 6.2: Splitting Input Vectors and Outputs / Labels of Training Data

```python
# Splitting Input Vectors and Outputs / Labels of Training Data

'''
*---------------- SPLIT_INPUT_VECTORS_AND_LABELS --------------*
|        Function: iloc()                                      |
|            Purpose: Splitting Input Vector and Labels        |
|        Arguments:                                            |
|            Attribute: Name or Location Attribute to Split    |
|        Return:                                               |
|            Attribute: Split Attributes                       |
*--------------------------------------------------------------*
'''

print("\n\nInputs Vectors (Feature Vectors) of Training Data:")
print("==================================================\n")
input_vector_train = training_data_encoded.iloc[: , :-1]
print(input_vector_train)

print("\n\nOutputs/Labels of Training Data:")
print("================================\n")
print("  Output")
output_label_train = training_data_encoded.iloc[: ,-1]
print(output_label_train)
```

## 6.3: Train the Support Vector Classifier

```python
# Train the Support Vector Classifier

''' 
*--------------- TRAIN_SUPPORT_VECTOR_CLASSIFIER ------------------*
|       Function: svm.SVC()                                        |
|           Purpose: Train the Algorithm on Training Data          |
|       Arguments:                                                 |
|           Training Data: Provide Training Data to the Model      |
|       Return:                                                    |
|           Parameter: Model return the Training Parameters        |
*------------------------------------------------------------------*
'''

print("\n\nTraining the Support Vector Classifier on Training Data")
print("========================================================\n")
print("\nParameters and their values:")
print("============================\n")
svc_model = svm.SVC(gamma='auto',random_state=0)
svc_model.fit(input_vector_train,np.ravel(output_label_train))
print(svc_model)
```

## Step 6.4: Save the Trained Model

```python
# Save the Trained Model

''' 
*--------------------- SAVE_THE_TRAINED_MODEL ---------------------*
|        Function: dump()                                          |
|             Purpose: Save the Trained Model on your Hard Disk    |
|        Arguments:                                                |
|             Model: Model Objects                                 |
|        Return:                                                   |
|             File: Trained Model will be Saved on Hard Disk       |
*------------------------------------------------------------------* 
'''

# Save the Model in a Pkl File

pickle.dump(svc_model, open('svc_trained_model.pkl', 'wb'))
```

# Step 7: Execute the Testing Phase 

## Step 7.1: Splitting Input Vectors and Outputs/Labels of Testing Data

```python
# Splitting Input Vectors and Outputs/Labels of Testing Data

'''
*---------------- SPLIT_INPUT_VECTORS_AND_LABELS --------------*
|        Function: iloc()                                      |
|            Purpose: Splitting Input Vector and Labels        |
|        Arguments:                                            |
|            Attribute: Name or Location Attribute to Split    |
|        Return:                                               |
|            Attribute: Split Attributes                       |
*--------------------------------------------------------------*
'''

print("\n\nInputs Vectors (Feature Vectors) of Testing Data:")
print("=================================================\n")
input_vector_test = testing_data_encoded.iloc[: , :-1]
print(input_vector_test)

print("\n\nOutputs/Labels of Testing Data:")
print("==============================\n")
print("  Output")
output_label_test = testing_data_encoded.iloc[: ,-1]
print(output_label_test)
```

## Step 7.2: Load the Saved Model

```python
# Load the Saved Model

''' 
*------------------- LOAD_SAVED_MODEL --------------------------*
|         Function: load()                                      |
|               Purpose: Method to Load Previously Saved Model  |
|         Arguments:                                            |
|               Model: Trained Model                            |
|         Return:                                               |
|               File: Saved Model will be Loaded in Memory      |
*---------------------------------------------------------------*
'''

# Load the Saved Model

model = pickle.load(open('svc_trained_model.pkl', 'rb'))
```

## Step 7.3: Evaluate the Machine Learning Model
### Step 7.3.1: Make Predictions with the Trained Models on Testing Data

```python
# Evaluate the Machine Learning Model

''' 
*--------------------- EVALUATE_MACHINE_LEARNING_MODEL ----------------------*
|       Function: Predict()                                                  |
|             Purpose: Make a Prediction using Algorithm on Test Data        |
|       Arguments:                                                           |
|            Testing Data: Provide Test data to the Trained Model            |
|       Return:                                                              |
|            Predictions: Model return Predictions                           |
*----------------------------------------------------------------------------* 
'''

# Provide Test data to the Trained Model

model_predictions = model.predict(input_vector_test)
testing_data_encoded.copy(deep=True)
pd.options.mode.chained_assignment = None
testing_data_encoded["Predictions"] = model_predictions

# Save the Predictions into CSV File

testing_data_encoded.to_csv(r'model-predictions.csv', index = False, header = True)

model_predictions = testing_data_encoded 
print("\n\nPredictions Returned by svc_trained_model:")
print("==========================================\n")
print(model_predictions)
```

## Step 7.4: Calculate the Accuracy Score

```python
# Calculate the Accuracy Score

''' 
/*------------------------ CALCULATE_ACCURACY_SCORE -------------------*
|          Function: accuracy_score()                                  |
|                Purpose: Evaluate the algorithm on Testing data       |
|          Arguments:                                                  |
|                Prediction: Predicted values                          |
|                Label: Actual values                                  |
|          Return:                                                     |
|                Accuracy: Accuracy Score                              |
*----------------------------------------------------------------------*
'''

# Calculate the Accuracy

model_accuracy_score = accuracy_score(model_predictions["Output"],model_predictions["Predictions"])

print("\n\nAccuracy Score:")
print("===============\n")
print(round(model_accuracy_score,2))
```

# Step 8: Execute the Application Phase

## Step 8.1: Take Input from User

```python
# Take Input from User

''' 
*---------------- TAKE_USER_INPUT ----------------*
'''

capshape_input = input("\nPlease enter Cap-shape here (bell=b,conical=c,convex=x,flat=f,knobbed=k,sunken=s) : ").strip()
capsurface_input = input("\nPlease enter your Cap-surface here (fibrous=f,grooves=g,scaly=y,smooth=s) : ").strip()
capcolor_input = input("\nPlease enter your Cap-color here (brown=n,buff=b,cinnamon=c,gray=g,green=r,pink=p,purple=u,red=e,white=w,yellow=y) : ").strip()
odor_input = input("\nPlease enter Odor here (almond=a,anise=l,creosote=c,fishy=y,foul=f,musty=m,none=n,pungent=p,spicy=s) : ").strip()
gillsize_input = input("\nPlease enter Gill-size here (broad=b,narrow=n) : ").strip()
gillcolor_input = input("\nPlease enter Gill-color here ( black=k,brown=n,buff=b,chocolate=h,gray=g,green=r,orange=o,pink=p,purple=u,red=e,white=w,yellow=y) : ").strip()
stalkroot_input = input("\nPlease enter Stal-kroot here (bulbous=b,club=c,cup=u,equal=e,rhizomorphs=z,rooted=r) : ").strip()
sporeprintcolor_input = input("\nPlease enter Spore-print-color here (black=k,brown=n,buff=b,chocolate=h,green=r,orange=o,purple=u,white=w,yellow=y) : ").strip()
population_input = input("\nPlease enter Population here (abundant=a,clustered=c,numerous=n,scattered=s,several=v,solitary=y) : ").strip()
habitat_input = input("\nPlease enter Habitat here (grasses=g,leaves=l,meadows=m,paths=p,urban=u,waste=w,woods=d) : ").strip()
```

## Step 8.2: Convert User Input into Feature Vector (Exactly Same as Feature Vectors of Sample Data)

```python
# Convert User Input into Feature Vector

user_input = pd.DataFrame({ 'cap-shape': [capshape_input],'cap-surface': [capsurface_input],'cap-color': [capcolor_input],
                           'odor': [odor_input],'gill-size':[gillsize_input],'gill-color':[gillcolor_input],'stalk-root':[stalkroot_input],
                           'spore-print-color':[sporeprintcolor_input],'population':[population_input],'habitat':[habitat_input]})

print("\n\nUser Input Feature Vector:")
print("==========================\n")
print(user_input)
```

## Step 8.3: Label Encoding of Feature Vector (Exactly Same as Label Encoded Feature Vectors of Sample Data)

```python
# Label Encoding

''' 
*----------------- TRANSFORM_UNSEEN_INTPUT_FEATURES ---------------*
|           Function: Transform()                                  |
|                 Purpose: Transform Input (Categorical) into      |
|                          Numerical Representation                |
|           Arguments:                                             |
|                 Attribute: Target values                         |
|           Return:                                                |
|                 Attribute: Numerical Representation              |
*------------------------------------------------------------------*   
'''

# Transform Input (Categorical) Attributes of Unseen Data into Numerical Representation

unseen_data_features = user_input.copy()
unseen_data_features["cap-shape"] =capshape_label_encoder.transform(user_input['cap-shape'])
unseen_data_features["cap-surface"] = capsurface_label_encoder.transform(user_input['cap-surface'])
unseen_data_features["cap-color"] = capcolor_label_encoder.transform(user_input['cap-color'])
unseen_data_features["odor"] = odor_label_encoder.transform(user_input['odor'])
unseen_data_features["gill-size"] = gillsize_label_encoder.transform(user_input['gill-size'])
unseen_data_features["gill-color"] =gillcolor_label_encoder.transform(user_input['gill-color'])
unseen_data_features["stalk-root"] = stalkroot_label_encoder.transform(user_input['stalk-root'])
unseen_data_features["spore-print-color"] = sporeprintcolor_label_encoder.transform(user_input['spore-print-color'])
unseen_data_features["population"] = population_label_encoder.transform(user_input['population'])
unseen_data_features["habitat"] = habitat_label_encoder.transform(user_input['habitat'])

print("\n\nUser Input Feature Vector:")
print("==========================\n")
print(user_input)

print("\n\nUser Input Encoded Feature Vector:")
print("==================================\n")
print(unseen_data_features)
```

## Step 8.4: Load the Saved Model

```python
# Load the Saved Model

''' 
*----------------------- LOAD_SAVED_MODEL --------------------------*
|         Function: load()                                          |
|             Purpose: Method to Load Previously Saved Model        |
|         Arguments:                                                |
|               Model: Trained Model                                |
|         Return:                                                   |
|               File: Saved Model will be Loaded in Memory          |
*-------------------------------------------------------------------*
'''

# Load the Saved Model

model = pickle.load(open('svc_trained_model.pkl', 'rb'))
```

## Step 8.5: Model Prediction
### Step 8.5.1: Apply Model on the Label Encoded Feature Vector of unseen instance and return Prediction to the User

```python
# Prediction of Unseen Instance

''' 
*----------------------------  MODEL_PREDICTION --------------------------*
|           Function: predict()                                           |
|                 Purpose: Use Trained Model to Predict the Output        |
|                          of Unseen Instances                            |
|           Arguments:                                                    |
|                 User Data: Label Encoded Feature Vector of              |
|                            Unseen Instances                             |
|           Return:                                                       |
|                 Survival: Survived or Not Survived                      |
*-------------------------------------------------------------------------*
'''

# Make a Prediction on Unseen Data

predicted_survival = model.predict(unseen_data_features)

if(predicted_survival == 1): 
    prediction = "Poisnous"
if(predicted_survival == 0):
    prediction = "Edible"

# Add the Prediction in a Pretty Table

pretty_table = PrettyTable()
pretty_table.add_column("       ** Prediction **       ",[prediction])
print(pretty_table)
```

# Step 9: Execute the Feedback Phase
## A Two-Step Process
